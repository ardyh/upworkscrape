```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

import re
import operator
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Firefox()
```

```python
INDONESIA_LINK = "indonesia"
USA_LINK = "united-states"
INDIA_LINK = "india"
UK_LINK = "united-kingdom"

CURRENT_LINK = USA_LINK
```

```python
el_found = True
current_page = 267
#all_freelancer_urls = []

while (el_found):
    if(current_page > 500):
        el_found = False
        break
    try:
        print("Page", current_page)
        
        driver.get("https://www.upwork.com/search/profiles/?revenue=1&pt=independent&loc="+CURRENT_LINK+"&page="+str(current_page))
        class_name = "freelancer-tile-name"
        freelancer_urls = merchant_links = WebDriverWait(driver, 10).until(
                                            EC.presence_of_all_elements_located((By.CLASS_NAME, class_name))
                                        )
        print("found", len(freelancer_urls), "freelancers")
        
        for freelancer in freelancer_urls:
            all_freelancer_urls.append([current_page, freelancer.get_attribute("href"), freelancer.get_attribute("title")])
        
        current_page += 1
        
    except Exception as e:
        try:
            #test if entered "verify you are a human" page
            driver.find_element_by_xpath("/html/body/section/div[2]/div/h1")
            driver = webdriver.Firefox()
            driver.implicitly_wait(5)
            print("new driver")
        except:
            print(e)
            el_found = False
```

```python
freelancer_urls = pd.DataFrame(all_freelancer_urls, columns=['page', 'link', 'name'])
freelancer_urls.drop_duplicates(subset="link", keep="first", inplace=True)
freelancer_urls.head()
print(freelancer_urls.shape)
```

```python
#assert 0 == 1
freelancer_urls.to_csv("../data/result/freelancer_urls_"+CURRENT_LINK+".csv", index=False)
```
